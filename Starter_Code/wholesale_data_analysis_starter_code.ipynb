## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df['subtotal'] = df['unit_price'] * df['qty']
df.head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
# gonna use lambda for this
df['total_weight'] = df['unit_weight'] * df['qty']
df['shipping_price'] = df['total_weight'].apply(lambda x: x * 7 if x>50 else x*10)
df.head()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
df['Line_price'] = round((df['subtotal'] + df['shipping_price'])*1.0925,2)
df.head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).

df['line_cost'] = df['unit_cost'] * df['qty'] + df['shipping_price']
df.head(3)
```

```python
# Create a column for the profit of each line using line cost and line price
df['line_profit'] = df['Line_price'] - df['line_cost']
df.head(3)
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
client_spend = {}
for id in top5_clients_list:
    total_price = df[df['client_id'] == id]['Line_price'].sum()
    client_spend[id] = total_price
    print(f"{id}: ${total_price:,.2f}")

    
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
client_ids = []
total_units_purchased = []
total_shipping_price = []
total_revenue =[]
total_profit = []

for id in top5_clients_list:
    cleint = df[df['client_id'] == id]
    client_ids.append(id)
    total_units_purchased.append(cleint['qty'].sum())
    total_shipping_price.append(cleint['shipping_price'].sum())
    total_revenue.append(cleint['Line_price'].sum())
    total_profit.append(cleint['line_profit'].sum())

summary_df = pd.DataFrame({ 'client id': client_ids, 'Total Units Purchased': total_units_purchased, 'Total Shipping Price': total_shipping_price,
                           'Total Revenue': total_revenue, 'Total Profit': total_profit})
summary_df
    
```

```python
money_columns = summary_df.drop('client id', axis = 1)
```

```python
# Format the data and rename the columns to names suitable for presentation.
# Define the money columns. 


# Define a function that converts a dollar amount to millions
def currency_format_millions(val):
    return val / 1_000_000
    
summary_df = summary_df.rename(columns=lambda x: x.replace(' (M) (M)', ' (M)'))
# Apply the currency_format_millions function to the money columns and create new columns for the millions format
for column in money_columns:
     if f'{column} (M)' not in summary_df.columns:
        summary_df[f'{column} (M)'] = summary_df[column].apply(currency_format_millions)



# Rename the columns to reflect the change in the money format
summary_df.rename(columns={
    summary_df.columns[0]: 'Client ID',
    summary_df.columns[1]: 'Total Units Purchased (Mills)',
    summary_df.columns[2]: 'Total Shipping Price (Mills)',
    summary_df.columns[3]: 'Total Revenue (Mills)',
    summary_df.columns[4]: 'Total Profit (Mills)'
}, inplace=True)

# Drop the original columns
summary_df
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
sorted_df = summary_df.sort_values(by='Total Profit (Mills)', ascending=False)
sorted_df = summary_df = summary_df.iloc[:, :5]
sorted_df
```
